```python
path='Players_overview'
files=os.listdir(path)
files=[name for name in files if name != ".DS_Store"]
rows=[]
for file in files:
    try:
        #name
        name=file.split('.')
        name=name[0]
        name=name.split('-')
        name=' '.join(name)
        #print(name)

        file=open(os.getcwd()+'/Players_overview/'+file,'r')
        e=file.read()
        soup = BeautifulSoup(e, 'html.parser')
        
        #position
        position=soup.find("div", attrs= {"class": "playerOverviewAside u-hide-mob"}).section.find_next().find_next()
        if position is None:
            position=None
        else:
            position=position.text
            position=position.strip()
            position=position.strip('\n')
        #print(position)
        
        #nationality
        nationality=soup.find("span", attrs= {"class": "playerCountry"})
        if nationality is None:
            nationality=None
        else:
            nationality=nationality.text
        #print(nationality)

        #age
        age=soup.find("ul",attrs={"class": "pdcol2"})
        if age is None:
            age is None
        else:
            age=age.li
            if age is None:
                age is None
            else:
                age=age.span
                if age is None:
                    dob=soup.find("ul",attrs={"class": "pdcol2"}).li.text
                    dob=dob.split('\n')
                    dob=dob[3]
                    dob=dob.split('/')
                    date=dob[0]
                    date=date.strip()
                    date=pd.to_numeric(date)
                    month=dob[1]
                    month=pd.to_numeric(month)
                    year=dob[2]
                    year=pd.to_numeric(year)
                    td=datetime.datetime.now().date()
                    bd=datetime.date(year,month,date)
                    age=int((td-bd).days /365.25)
                else:
                    age=age.text
                    age=age.split('(')
                    age=age[1]
                    age=age.split(')')
                    age=age[0]
        #print(age)
    
        #height
        height=soup.find("ul",attrs={"class": "pdcol3"})
        if height is None:
            height is None
        else:
            height=height.li
            if height is None:
                height=None
            else:
                height=height.text
                height=height.split("\n")
                height=height[2]
        #print(height)

        row=[name,position,nationality,age,height]
        rows.append(row)
        
    except:
        print("error")
        print(file)
        continue
```
